```python
#Cambiamos el tipo de dato de las columnas que contienen fechas
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'], errors='coerce')
```

```python
#Eliminamos el valor nan de la columna target_audience
df = df.dropna(subset=['target_audience'])

#Miramos como queda la columna target_audience despues del cambio
df['target_audience'].value_counts(dropna=False)
```

```python
#Unificamos y corregimos lo que haty en la columan type
df['type'] = df['type'].str.lower().str.strip()

df['type'] = df['type'].replace({
    'email ': 'email',  
    'webina': 'webinar',  
    'socialmedia': 'social media'  
})

#Miramos como queda la columna type despues del cambio
print(df['type'].value_counts())
```

```python
#Normalizamos todas las columnas 
df = df.apply(lambda col: col.str.lower().str.strip() if col.dtype == "object" else col)

print(df.head())
```

```python
#Añadimos variables derivadas
df['net_profit'] = df['revenue'] - df['budget'] #Calculamos el beneficio neto
print(df[['revenue', 'budget', 'net_profit']].head())
```

```python
#Extraemos componentes temporales de la fecha
df['start_year'] = df['start_date'].dt.year
df['start_month'] = df['start_date'].dt.month
df['start_quarter'] = df['start_date'].dt.quarter

df['end_year'] = df['end_date'].dt.year
df['end_month'] = df['end_date'].dt.month
df['end_quarter'] = df['end_date'].dt.quarter


print(df[['start_date', 'start_year', 'start_month', 'start_quarter', 'end_date', 'end_year', 'end_month', 'end_quarter']].head())
```

```python
#Compruebo si el paso anterior esta bien con un ejemplo
print(df[['campaign_name', 'start_date', 'start_year', 'start_month', 'start_quarter', 'end_date', 'end_year', 'end_month', 'end_quarter']].head())
```

```python
#Creamos categorias de rendimiento basadas en roi y tasa de conversion
def asignar_categoria(row):  
    if row['roi'] <= 0.3: #Roi categorizado
        roi_categoria = 'Bajo'
    elif row['roi'] <= 0.6:
        roi_categoria = 'Medio'
    else:
        roi_categoria = 'Alto'
    
    if row['conversion_rate'] <= 0.2:  #Tasa de conversion con categorias
        conversion_categoria = 'Baja'
    elif row['conversion_rate'] <= 0.5:
        conversion_categoria = 'Media'
    else:
        conversion_categoria = 'Alta'
    
    if roi_categoria == 'Alto' and conversion_categoria == 'Alta':  #Asignacion de categorias
        return 'Alto rendimiento'
    elif roi_categoria == 'Bajo' and conversion_categoria == 'Baja':
        return 'Bajo rendimiento'
    else:
        return 'Rendimiento medio'

df['categoria_rendimiento'] = df.apply(asignar_categoria, axis=1)  #Creamos la columna performance_category


print(df[['campaign_name', 'roi', 'conversion_rate', 'categoria_rendimiento']].head())
```

```python
#Calculo de metricas de eficiencia
df['cost_per_conversion'] = df['budget'] / df['conversion_rate'] #Coste por conversion

print(df[['campaign_name', 'budget', 'conversion_rate', 'cost_per_conversion']].head())
```

```python
#Variables binarias para sacar informacion importante 

#Consideramos los siguientes umbrales para las variables binarias:
umbral_bajo_presupuesto = 10000  #Presupuesto es menor a 10,000
umbral_alto_roi = 0.6  #ROI es mayor a 0.6
umbral_alta_conversion = 0.5  #Tasa de conversión es mayor a 0.5
umbral_duracion_larga = pd.Timedelta(days=180)  #Duración mayor a 6 mese = 180 dias

#Flag para presupuesto bajo
df['low_budget_flag'] = (df['budget'] < umbral_bajo_presupuesto).astype(int)

#Flag para alto ROI
df['high_roi_flag'] = (df['roi'] > umbral_alto_roi).astype(int)

#Flag para alta tasa de conversión
df['high_conversion_flag'] = (df['conversion_rate'] > umbral_alta_conversion).astype(int)

#Flag para "pago"
df['paid_channel_flag'] = (df['channel'] == 'paid').astype(int)

#Flag para duración larga de la campaña
df['long_duration_flag'] = (df['end_date'] - df['start_date'] > umbral_duracion_larga).astype(int)

#Flag para B2B 
df['b2b_flag'] = (df['target_audience'] == 'B2B').astype(int)


print(df[['campaign_name', 'budget', 'roi', 'conversion_rate', 'channel', 'start_date', 'end_date',
          'low_budget_flag', 'high_roi_flag', 'high_conversion_flag', 'paid_channel_flag', 
          'long_duration_flag', 'b2b_flag']].head())
```
