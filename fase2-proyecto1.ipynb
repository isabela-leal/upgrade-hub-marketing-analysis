```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargamos nuestro dataset limpio
df = pd.read_csv('C:/Users/Propietario/OneDrive/Escritorio/DataAnalysis2025/Proyecto-1/marketingcampaigns_cleaned.csv')
```

Â¿Que canal de marketing se utiliza con mayor frecuencia y cual genera mejor ROI?


```python
#Visualizamos la frecuencia de los canales de marketing
channel_counts = df['channel'].value_counts()

# Creacion del grafico de la frecuencia de los canales de marketing
plt.figure(figsize=(10,6))
sns.barplot(x=channel_counts.index, y=channel_counts.values, palette="Blues")
plt.title('Frecuencia de Uso de Canales de Marketing', fontsize=14)
plt.xlabel('Canal de Marketing')
plt.ylabel('NÃºmero de CampaÃ±as')
plt.xticks(rotation=45)
plt.show()

# Comparamos el rendimiento de cada canal en tÃ©rminos de ROI
plt.figure(figsize=(12,6))
sns.boxplot(x='channel', y='roi', data=df, palette="Blues")
plt.title('DistribuciÃ³n del ROI por Canal de Marketing', fontsize=14)
plt.xlabel('Canal de Marketing')
plt.ylabel('ROI')
plt.xticks(rotation=45)
plt.show()

# ComparaciÃ³n de ROI por canal segmentado por tipo de campaÃ±a
plt.figure(figsize=(12,6))
sns.boxplot(x='channel', y='roi', hue='type', data=df, palette="Blues")
plt.title('DistribuciÃ³n del ROI por Canal y Tipo de CampaÃ±a', fontsize=14)
plt.xlabel('Canal de Marketing')
plt.ylabel('ROI')
plt.xticks(rotation=45)
plt.show()

# ComparaciÃ³n de ROI por canal segmentado por 'target_audience'
plt.figure(figsize=(12,6))
sns.boxplot(x='channel', y='roi', hue='target_audience', data=df, palette="Blues")
plt.title('DistribuciÃ³n del ROI por Canal y Audiencia Objetivo', fontsize=14)
plt.xlabel('Canal de Marketing')
plt.ylabel('ROI')
plt.xticks(rotation=45)
plt.show()
```

Â¿Que tipo de campaÃ±a genera mas ingresos en promedio y cual tiene mejor converision?

```python
#Calculamos ls ingresos promedio y la tasa de conversiÃ³n promedio por tipo de campaÃ±a
campaign_metrics = df.groupby('type').agg(
    average_revenue=('revenue', 'mean'),
    average_conversion_rate=('conversion_rate', 'mean')
).reset_index()

#Creamos grafico para los ingresos promedio por tipo de campaÃ±a
plt.figure(figsize=(12,6))
sns.barplot(x='type', y='average_revenue', data=campaign_metrics, palette="Blues")
plt.title('Ingresos Promedio por Tipo de CampaÃ±a', fontsize=14)
plt.xlabel('Tipo de CampaÃ±a')
plt.ylabel('Ingresos Promedio')
plt.xticks(rotation=45)
plt.show()

# Visualizamos la tasa de conversiÃ³n promedio por tipo de campaÃ±a
plt.figure(figsize=(12,6))
sns.barplot(x='type', y='average_conversion_rate', data=campaign_metrics, palette="Blues")
plt.title('Tasa de ConversiÃ³n Promedio por Tipo de CampaÃ±a', fontsize=14)
plt.xlabel('Tipo de CampaÃ±a')
plt.ylabel('Tasa de ConversiÃ³n Promedio')
plt.xticks(rotation=45)
plt.show()

#CorrelaciÃ³n entre ingresos y tasa de conversiÃ³n 
plt.figure(figsize=(10,6))
sns.scatterplot(x='average_revenue', y='average_conversion_rate', data=campaign_metrics, palette="Blues", s=100)
plt.title('CorrelaciÃ³n entre Ingresos Promedio y Tasa de ConversiÃ³n por Tipo de CampaÃ±a', fontsize=14)
plt.xlabel('Ingresos Promedio')
plt.ylabel('Tasa de ConversiÃ³n Promedio')
plt.show()
```

Â¿Hay diferencias en la tasa de conversion entre audiencias B2B Y B2C?

```python
#Boxplot comparativo entre la tasa de conversiÃ³n de B2B y B2C
plt.figure(figsize=(10,6))
sns.boxplot(x='target_audience', y='conversion_rate', data=df, palette="Blues")
plt.title('ComparaciÃ³n de Tasa de ConversiÃ³n entre Audiencias B2B y B2C', fontsize=14)
plt.xlabel('Audiencia')
plt.ylabel('Tasa de ConversiÃ³n')
plt.show()

#Promedio de la tasa de conversiÃ³n por audiencia, canal y tipo de campaÃ±a
audience_conversion = df.groupby(['target_audience', 'channel']).agg(
    avg_conversion_rate=('conversion_rate', 'mean')
).reset_index()

#Creamos su respectivo grafico
plt.figure(figsize=(12,6))
sns.barplot(x='channel', y='avg_conversion_rate', hue='target_audience', data=audience_conversion, palette="Blues")
plt.title('Tasa de ConversiÃ³n Promedio por Audiencia y Canal', fontsize=14)
plt.xlabel('Canal')
plt.ylabel('Tasa de ConversiÃ³n Promedio')
plt.xticks(rotation=45)
plt.show()

#Promedio de la tasa de conversiÃ³n por audiencia y tipo de campaÃ±a
audience_campaign_conversion = df.groupby(['target_audience', 'type']).agg(
    avg_conversion_rate=('conversion_rate', 'mean')
).reset_index()

#Y sacamos el grafico
plt.figure(figsize=(12,6))
sns.barplot(x='type', y='avg_conversion_rate', hue='target_audience', data=audience_campaign_conversion, palette="Blues")
plt.title('Tasa de ConversiÃ³n Promedio por Audiencia y Tipo de CampaÃ±a', fontsize=14)
plt.xlabel('Tipo de CampaÃ±a')
plt.ylabel('Tasa de ConversiÃ³n Promedio')
plt.xticks(rotation=45)
plt.show()
```

Â¿Que campaÃ±a tiene el mayor beneficio neto (net_profit)? Â¿Que caracteristicas la hacen exitosa?

```python
#Calcular beneficio neto y ordenar por beneficio neto
df['net_profit'] = df['revenue'] - df['budget']

#Ordenar las campaÃ±as por beneficio neto 
top_campaigns = df[['campaign_name', 'net_profit', 'budget', 'revenue', 'roi', 'conversion_rate', 'type', 'target_audience', 'channel']].sort_values(by='net_profit', ascending=False).head(10)

# Miramos las 10 campaÃ±as con mayor beneficio nneto
print(top_campaigns[['campaign_name', 'net_profit', 'budget', 'revenue', 'roi', 'conversion_rate', 'type', 'target_audience', 'channel']])

# Empleamos un grÃ¡fico de barras de beneficio neto para las top 10 campaÃ±as
plt.figure(figsize=(12,6))
sns.barplot(x='net_profit', y='campaign_name', data=top_campaigns, palette="Blues")
plt.title('Top 10 CampaÃ±as por Beneficio Neto', fontsize=14)
plt.xlabel('Beneficio Neto')
plt.ylabel('CampaÃ±a')
plt.show()

#GrÃ¡fico de barras comparativo de caracterÃ­sticas (canal, tipo, audiencia) de las top 10 campaÃ±as. Analizamos la distribuciones

# DistribuciÃ³n por canal
plt.figure(figsize=(12,6))
sns.countplot(y='channel', data=top_campaigns, palette="Blues")
plt.title('DistribuciÃ³n de Canales en las Top 10 CampaÃ±as por Beneficio Neto', fontsize=14)
plt.xlabel('NÃºmero de CampaÃ±as')
plt.ylabel('Canal')
plt.show()

# DistribuciÃ³n por tipo de campaÃ±a
plt.figure(figsize=(12,6))
sns.countplot(y='type', data=top_campaigns, palette="Blues")
plt.title('DistribuciÃ³n de Tipos de CampaÃ±as en las Top 10 CampaÃ±as por Beneficio Neto', fontsize=14)
plt.xlabel('NÃºmero de CampaÃ±as')
plt.ylabel('Tipo de CampaÃ±a')
plt.show()

# DistribuciÃ³n por audiencia objetivo
plt.figure(figsize=(12,6))
sns.countplot(y='target_audience', data=top_campaigns, palette="Blues")
plt.title('DistribuciÃ³n de Audiencias en las Top 10 CampaÃ±as por Beneficio Neto', fontsize=14)
plt.xlabel('NÃºmero de CampaÃ±as')
plt.ylabel('Audiencia Objetivo')
plt.show()
```

Â¿Existen patrones estacionales o temporales en el rendimiento de las campaÃ±as?

```python
#Componentes temporales 
#Extraemos componentes temporales de la fecha
df['start_date'] = pd.to_datetime(df['start_date'])  # Asegurarnos de que 'start_date' sea de tipo datetime

df['month'] = df['start_date'].dt.month  # Extraer el mes
df['quarter'] = df['start_date'].dt.quarter  # Extraer el trimestre
df['year'] = df['start_date'].dt.year  # Extraer el aÃ±o

df[['start_date', 'month', 'quarter', 'year']].head()
```

```python
#Creamos grafico de linea temporalÃ± del rendimiento de las campaÃ±as

#Agrupar por aÃ±o y mes para calcular la media de ROI e ingresos
monthly_performance = df.groupby(['year', 'month']).agg({'roi': 'mean', 'revenue': 'mean'}).reset_index()

#GrÃ¡fico de lÃ­nea temporal para analizar el rendimiento a lo largo del tiempo
plt.figure(figsize=(12, 6))

#GrÃ¡fico de ROI mensual
sns.lineplot(data=monthly_performance, x='month', y='roi', marker='o', label='ROI', color='blue')

# GrÃ¡fico de Ingresos mensuales
sns.lineplot(data=monthly_performance, x='month', y='revenue', marker='o', label='Ingresos', color='green')

# Damos forma al grafico
plt.title('Rendimiento mensual de las campaÃ±as (ROI e Ingresos)', fontsize=16)
plt.xlabel('Mes', fontsize=14)
plt.ylabel('Valor', fontsize=14)
plt.legend(title='MÃ©tricas', fontsize=12)
plt.grid(True)
plt.xticks(range(1, 13))  # Aseguramos que los meses estÃ©n bien representados
plt.show()
```

```python
#Mapa de calor para el rendimiento mensual por tipo de campaÃ±a
#Agrupar por aÃ±o, mes y tipo de campaÃ±a para obtener el rendimiento 
monthly_campaign_performance = df.groupby(['year', 'month', 'type']).agg({'roi': 'mean', 'revenue': 'mean'}).reset_index()

#Pivoteamos los datos para crear un formato adecuado para un heatmap 
heatmap_data = monthly_campaign_performance.pivot_table(index='month', columns='type', values='roi', aggfunc='mean')

#Creacion de mapa de calor
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='Blues', fmt='.2f', cbar_kws={'label': 'Promedio ROI'}, linewidths=0.5)

#Damos la forma que queremos a nuestro grafico
plt.title('Heatmap del ROI mensual por tipo de campaÃ±a', fontsize=16)
plt.xlabel('Tipo de CampaÃ±a', fontsize=14)
plt.ylabel('Mes', fontsize=14)
plt.show()
```

```python
#Analisis de rendimiento por trimestre
#Agrupar por aÃ±o y trimestre para analizar el rendimiento por trimestre
quarterly_performance = df.groupby(['year', 'quarter']).agg({'roi': 'mean', 'revenue': 'mean'}).reset_index()

#GrÃ¡fico de lÃ­nea temporal por trimestre para ver la tendencia
plt.figure(figsize=(12, 6))

#GrÃ¡fico de ROI trimestral
sns.lineplot(data=quarterly_performance, x='quarter', y='roi', marker='o', label='ROI', color='blue')

#GrÃ¡fico de Ingresos trimestrales
sns.lineplot(data=quarterly_performance, x='quarter', y='revenue', marker='o', label='Ingresos', color='green')

#Terminamios nuestro grafico
plt.title('Rendimiento trimestral de las campaÃ±as (ROI e Ingresos)', fontsize=16)
plt.xlabel('Trimestre', fontsize=14)
plt.ylabel('Valor', fontsize=14)
plt.legend(title='MÃ©tricas', fontsize=12)
plt.grid(True)
plt.show()
```
